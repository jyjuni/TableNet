### Train OCR With Tesstrain

We use tesstrain module provided by Tesseract Project to train our OCR model: https://github.com/tesseract-ocr/tesstrain

The tesstrain module is on command-line with maketools only. We provide our training commands as follows. The codes are compatible for Tesseract 4 and Tesseract 5. 

<b>Note:</b> Please replace `~/eyegaze/tesstrain` with your tesstrain directory and replace `/home/linuxbrew/.linuxbrew/share/tessdata/` with your tessdata directory accordingly. 

```python
# copy traindata
!cp -r dataset/Fintabnet_data/ocr_traindata/cell  ~/eyegaze/tesstrain/data/ocr_traindata

# train with tesstrain
!cd ~/eyegaze/tesstrain/ && make training MODEL_NAME=foo DATA_DIR=data GROUND_TRUTH_DIR=data/ocr_traindata 

# copy trained model from tesstrain to our project dataset: 
!cp ~/eyegaze/tesstrain/data/foo.traineddata dataset/foo.traineddata 

# add to tesseract langugage models 
!sudo cp dataset/ocr_full.traineddata /home/linuxbrew/.linuxbrew/share/tessdata/fintabnet_full.traineddata #REPLACE BY YOUR TESSERACT TESSDATA DIRECTORY
```
